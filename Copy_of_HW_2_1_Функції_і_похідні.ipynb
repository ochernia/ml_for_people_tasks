1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 10 * x - 5
```

```python
def df(x):
    if isinstance(x, int):
        return 10
    else: 
        return [10] * len(x)
```

```python
x = np.linspace(0,10,20)

plt.figure(figsize=(10, 5))
plt.plot(x, f(x), label='f(x) = 10x-5', color='green')
plt.plot(x, df(x), label="f'(x) = 10", color='red')

plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
def f_sin(x):
    return np.sin(x)

def df_sin(x):
    return np.cos(x)

def f_ln(x):
    return np.log(x)

def df_ln(x):
    return 1 / x
```

```python
x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x_ln = np.linspace(0.1, 10, 500)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_sin, f_sin(x_sin), label='f(x) = sin(x)', color='green')
plt.plot(x_sin, df_sin(x_sin), label="f'(x) = cos(x)", color='red')
plt.title('sin(x) and its derivative Python')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x_ln, f_ln(x_ln), label='f(x) = ln(x)', color='green')
plt.plot(x_ln, df_ln(x_ln), label="f'(x) = 1/x", color='red')
plt.title('ln(x) and its derivative Python')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import sympy as sp
```

```python
x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)


df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)

df_sin, df_ln
```

```python
f_sin_func = sp.lambdify(x, f_sin, 'numpy')
df_sin_func = sp.lambdify(x, df_sin, 'numpy')

f_ln_func = sp.lambdify(x, f_ln, 'numpy')
df_ln_func = sp.lambdify(x, df_ln, 'numpy')

x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x_ln = np.linspace(0.1, 10, 500)

# Plot sin(x) and its derivative
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_sin, f_sin_func(x_sin), label='f(x) = sin(x)', color='green')
plt.plot(x_sin, df_sin_func(x_sin), label="f'(x) = cos(x)", color='red')
plt.title('sin(x) and its derivative SymPy')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Plot ln(x) and its derivative
plt.subplot(1, 2, 2)
plt.plot(x_ln, f_ln_func(x_ln), label='f(x) = ln(x)', color='green')
plt.plot(x_ln, df_ln_func(x_ln), label="f'(x) = 1/x", color='red')
plt.title('ln(x) and its derivative SymPy')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.Symbol('x')
f = x**2 - 5*x + 1

df = sp.diff(f, x)

df
```

```python
f_numeric = sp.lambdify(x, f, 'numpy')
df_numeric = sp.lambdify(x, df, 'numpy')

x_values = np.linspace(0, 100)
f_values = f_numeric(x_values)
df_values = df_numeric(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x) = x^2 - 5x + 1', color='green')
plt.plot(x_values, df_values, label="f'(x) = 2x - 5", color='red')
plt.title('Функція загальних витрат компанії та її похідна')
plt.xlabel('x (Кількість вироблених одиниць)')
plt.ylabel('y (Витрати)')
plt.legend()
plt.grid(True)

plt.show()
```
